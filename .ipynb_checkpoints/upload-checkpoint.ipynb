```python
import os
import shutil
import time


path= '/Users/zhangjifeng/Documents/android/workspace/Translate_Pen/'

os.chdir(path)

new_version_text = '''<a href="http://106.120.188.43/app/tr2/android/%s" class="list-group-item">
<h4 class="list-group-item-heading">%s 点击下载</h4>
<p class="list-group-item-text">
%s
</p>
</a>
'''
    
    
def rm_apk():
    os.chdir(os.path.join(path, 'dailybuild'))
    for f in os.listdir():
        if f.endswith('.apk'):
            os.remove(f)
            print("del", f)
            
def build():
    os.chdir(path)
    os.system('./gradlew clean assembleRelease')
    os.chdir(os.path.join(path, 'app/build/outputs/apk/release/'))
    apk = ""
    for f in os.listdir():
        if f.endswith('.apk'):
            apk = f
            shutil.copy(f, os.path.join(path, 'dailybuild/'+f))
    return apk


def writeto(line):
    with open('index1.html','a') as f:
        f.write(line)
        print(line)

def construct(apk, changelog):
    if len(apk)==0:
        print('apk not found')
        return
    curtime = time.strftime('%Y年%m月%d日 %H:%M', time.localtime())
    text = new_version_text % (apk, curtime, changelog)
    print(text)
    need_to_write = True
    with open('index.html', 'r') as f1:
        for line in f1.readlines():
            if need_to_write and line.startswith('<a href="http://106.120.188.43/app/tr2'):
                writeto(text)
                need_to_write = False
            writeto(line) 

def scp(apk):
    os.chdir(os.path.join(path,'dailybuild'))
    command1 = 'sshpass -p "noSafeNoWork@2014" scp %s root@10.152.232.141:/search/m2/app/tr2/android/%s' % (apk, apk) 
    print(command1)
    os.system(command1)
    command2 = 'sshpass -p "noSafeNoWork@2014" scp index.html root@10.152.232.141:/search/m2/app/tr2/android/index.html'
    print(command2)
    os.system(command2)
            
apkname = build()
print(apkname)
os.chdir(os.path.join(path, 'dailybuild'))
shutil.copy('index.html','index_bak.html')
if os.path.exists('index1.html'):
    os.remove('index1.html')
construct(apkname, "1.对话翻译</br>2.分享</br>3.波形图</br>4.连接优化")
shutil.move('index1.html', 'index.html')
scp(apkname)
rm_apk()
```
