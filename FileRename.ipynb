```python
from os import listdir
from os.path import isfile, join

path = '/Users/zhangjifeng/Documents/天气'
files = [f for f in listdir(path) if isfile(join(path, f)) and not f.startswith('.')]

files_3 = [f for f in files if '@3×' in f]
files_3
```

```python

def transformStr(str):
    
    ret = str.replace('Copy 2@3×','icon')
    ret = ret.replace('Copy 3@3×','icon')
    ret = ret.replace('Copy@3×','icon')
    ret = ret.replace('copy 2@3×','icon')
    ret = ret.replace('copy 3@3×','icon')
    ret = ret.replace('copy@3×','icon')
    ret = ret.replace('@3×','bg')
    ret = ret.replace(' ','_')
    ret = ret.replace('__','_')
    ret = ret.replace('__','_')
    splited = ret.split('_')
    splited.remove(splited[1])
    ret = '_'.join(s for s in splited)
    ret = ret.lower()
    return ret


cleaned_files = [transformStr(f) for f in files_3 if not f.startswith('.')]

for index in range(len(cleaned_files)):
    print('http://s6.tm.sogoucdn.com/static/tr/headset/weather/'+cleaned_files[index])
```
